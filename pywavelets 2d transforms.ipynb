```python
from __future__ import division
import numpy as np
import scipy
import skimage
import matplotlib.pyplot as plt
%matplotlib inline
from skimage import data, io, filters
import pywt
```

```python
im = data.camera()
io.imshow(im,  cmap = 'Greys_r')
```

```python
transform = pywt.wavedec2(im, 'db2')
recovery = pywt.waverec2(transform, 'db2')
```

```python
io.imshow(recovery - im, cmap='Greys_r') #i would expect this to give me a blank image (all zeros)
print (np.sum(np.ravel((recovery - im)**2)))**(0.5) #and this to be ~0
```
